## All imports and static HTML website prep


```python
from requests import get
url = 'https://biopharmguy.com/links/company-by-name-northern-california.php'
response = get(url)
print(response.text[:500])
import lxml.html as lh
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
html_soup = BeautifulSoup(response.text, 'html.parser')
type(html_soup)
```

## Initial exploration of data

```python
page = get(url)
doc = lh.fromstring(page.content)
tr_elements = doc.xpath('//tr')
print (len(tr_elements))
tr_elements[3].text_content()
```

## Create empty list

```python
number = np.arange(0,1020,1)
df = pd.DataFrame(index=number, columns=['Name','Location','Description'])
df
```

## Read from HTML and fill DataFrame

```python
table = html_soup.find_all('table')[0]

row_marker = 0
for row in table.find_all('tr'):
    column_marker = 0
    columns = row.find_all('td')
    for column in columns:
        df.iat[row_marker,column_marker] = column.get_text()
        column_marker += 1
    if len(columns) > 0:
        row_marker += 1
df
```
